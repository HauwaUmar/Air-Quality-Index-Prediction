```python
import pandas as pd
```

```python
air_index = pd.read_csv('pollution_2000_2021.csv')
temp_2019 = pd.read_csv('daily_TEMP_2019.csv')
temp_2020 = pd.read_csv('daily_TEMP_2020.csv')
temp_2021 = pd.read_csv('daily_TEMP_2021.csv')
temp_2022 = pd.read_csv('daily_TEMP_2022.csv')
```

```python
means = final_df.groupby(['State'])['Temperature'].mean()
final_df = final_df.set_index(['State'])
final_df['Temperature'] = final_df['Temperature'].fillna(means)
final_df = final_df.reset_index()
print(final_df)
```

```python

# importing Randomforest
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split


# separating class label and other attributes
train1 = final_df.drop(['AQI','State','Date'], axis=1)
target = final_df['AQI']
X_train, X_test, y_train, y_test = train_test_split(train1, target, test_size=0.2, random_state=5)

 
# creating model
m1 = RandomForestRegressor()
 

 
# Fitting the model
m1.fit(X_train, y_train)
'''RandomForestRegressor(bootstrap=True, ccp_alpha=0.0, criterion='mse',
                      max_depth=None, max_features='auto', max_leaf_nodes=None,
                      max_samples=None, min_impurity_decrease=0.0,
                      min_impurity_split=None, min_samples_leaf=1,
                      min_samples_split=2, min_weight_fraction_leaf=0.0,
                      n_estimators=100, n_jobs=None, oob_score=False,
                      random_state=None, verbose=0, warm_start=False)'''
 
# calculating the score and the score is  97.96360799890066%
m1.score(X_test, y_test) * 100
 
```

```python
import pandas as pd

dataset = pd.read_csv('pollution_2000_2021.csv')
```

```python
dataset.info()
```

```python
dataset = dataset.drop(['City','Address','County'], axis=1)
```

```python
dataset = dataset.drop(['State'], axis=1)
pollutants = ['no2_aqi', 'o3_aqi', 'so2_aqi', 'co_aqi']

grouped_data = dataset.groupby(['Date']).mean()
grouped_data.index = pd.to_datetime(grouped_data.index)
grouped_data.head()
```

```python
grouped_data.rename(columns = {'O3 AQI':'o3_aqi', 'CO AQI':'co_aqi','SO2 AQI':'so2_aqi','NO2 AQI':'no2_aqi'}, inplace = True)
```

```python
grouped_data
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
no2_dec = seasonal_decompose(grouped_data['no2_aqi'], model='additive', freq=365)
o3_dec = seasonal_decompose(grouped_data['o3_aqi'], model='additive', freq=365)
so2_dec = seasonal_decompose(grouped_data['so2_aqi'], model='additive', freq=365)
co_dec = seasonal_decompose(grouped_data['co_aqi'], model='additive', freq=365)
```

```python
train = grouped_data[:'2018-01-01']
test = grouped_data['2018-01-01':'2021-01-01']
```

```python
train.shape
```

```python
test.shape
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import RandomForestRegressor
 
# creating model
m1 = RandomForestRegressor()
 
# separating class label and other attributes
train1 = train.drop(['co_aqi'], axis=1)
target = train['co_aqi']
 
# Fitting the model
m1.fit(train1, target)
'''RandomForestRegressor(bootstrap=True, ccp_alpha=0.0, criterion='mse',
                      max_depth=None, max_features='auto', max_leaf_nodes=None,
                      max_samples=None, min_impurity_decrease=0.0,
                      min_impurity_split=None, min_samples_leaf=1,
                      min_samples_split=2, min_weight_fraction_leaf=0.0,
                      n_estimators=100, n_jobs=None, oob_score=False,
                      random_state=None, verbose=0, warm_start=False)'''
 
# calculating the score and the score is  97.96360799890066%
m1.score(train1, target) * 100
 
# predicting the model with other values (testing the data)
# so AQI is 123.71

# Adaboost model
# importing module
 
# defining model
m2 = AdaBoostRegressor()
 
# Fitting the model
m2.fit(train1, target)
 
'''AdaBoostRegressor(base_estimator=None, learning_rate=1.0, loss='linear',
                  n_estimators=50, random_state=None)'''
 
# calculating the score and the score is  96.15377360010211%
m2.score(train1, target)*100
 
# predicting the model with other values (testing the data)
# so AQI is 94.42105263
```

```python
a = test.drop(['co_aqi'], axis=1)
b= test['co_aqi']
mod_pred_1 = m1.score(a,b)*100

mod_pred_2 = m2.score(a,b)*100
```
